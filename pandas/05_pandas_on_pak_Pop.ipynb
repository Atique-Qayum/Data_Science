```python
# how much rural population of Pakistan is increased from 1998 to 2017?
p2017 = df['ALL SEXES (RURAL)'].sum()
p1998 = df['POPULATION 1998 (RURAL)'].sum()

difference = (p2017 - p1998)
difference
```

```python
print("The total rural population of Pakistan have increase by ", difference)
```
