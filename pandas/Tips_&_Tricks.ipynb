# split dataset into tow subsets

```python
df = pd.read_excel('titanic.xlsx')
ship_1 = df.sample(frac=0.5, random_state=1)
ship_1.shape
```

```python
ship_2 = df.drop(ship_1.index)
ship_2.shape
```

```python
ship_1.head()
```

```python
ship_2.head()
```

```python
df.shape
```

```python
# hence the data is split into tow subsets
len(ship_1)+len(ship_2)
```

# joining tow subsets into one 

```python
df1 = ship_1.append(ship_2)
df1.shape
```
